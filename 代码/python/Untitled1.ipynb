```python
# -*- coding:utf-8 -*-
class Solution:
    def movingCount(self, threshold, rows, cols):
        def sub(list):
            res = 0
            for i in list:
                res = res + i
            return res
        # write code here
        cout = 0
        for i in range(rows-1):
            for j in range(cols-1):
                i_list = [int(n) for n in str(i)]
                j_list = [int(n) for n in str(j)]
                res = sub(i_list) + sub(j_list)
                if res <= threshold:
                    cout = cout + 1
        return cout
```

```python
# -*- coding:utf-8 -*-
class Solution:
    def movingCount(self, threshold, rs, cs):
        mark = [False]*(rs*cs)
        count = self.getCount(threshold,rs,cs,0,0,mark)
        return count
    
    
    def getCount(self,threshold,rs,cs,r,c,mark):
        count = 0
        if self.getRes(threshold,rs,cs,r,c,mark):
            mark[r * cs + c] = True
            count = 1 + self.getCount(threshold,rs,cs,r,c + 1,mark) + self.getCount(threshold,rs,cs,r,c - 1,mark) + self.getCount(threshold,rs,cs,r + 1,c,mark) + self.getCount(threshold,rs,cs,r - 1,c,mark) 
        return count
    
    def getRes(self,threshold,rs,cs,r,c,mark):
        if self.getSum(r) + self.getSum(c) <= threshold and not mark[r * cs + c] and r < rs and c < cs and r >= 0 and c >= 0:
            return True
        return False
    
    def getSum(self,number):
        sumNum = 0
        while number > 0:
            sumNum += number % 10
            number = number // 10
        return sumNum
            
            
```

```python
# -*- coding:utf-8 -*-
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    # 返回从尾部到头部的列表值序列，例如[1,2,3]
    def printListFromTailToHead(self, listNode):
        # write code here
        if listNode is None:
            return []
        res = []
        while listNode is not None:
            res.extend([listNode.val])
            listNode = listNode.next
        res.extend([listNode.val])
        return res[::-1]
            
```
